Rename column Names



```python
data.columns=['States','Data','Frequency','Estimated unemploymnet Rate','Estimated Employed','Estimated labour participation rate','Region','longitude','Latitude']
```

Estimated no.of employee according to different regions of india

```python
data.columns=['States','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate','Region','longitude','latitude']
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed',hue='Region',data=data)
plt.show()
```

Unemployment rate according to different regions of india

```python
plt.figure(figsize=(10,8))
plt.title("Indian Unemployment")
sns.histplot(x='Estimated Unemployment Rate',hue='Region',data=data)
plt.show()
```

Dashboard to analyze the unemployment rate of each indian state

```python
import plotly.express as px

unemployment = data[['States', 'Region', 'Estimated Unemployment Rate']]

figure = px.sunburst(
    unemployment,
    path=['Region', 'States'],
    values='Estimated Unemployment Rate',
    width=700,
    height=600,
    color='Estimated Unemployment Rate',
    color_continuous_scale='Viridis',
    title='Unemployment Rate in India'
)
figure.show()
```

Pie Chart for Region-Wise Data Distribution

```python
region_counts = data['Region'].value_counts()
plt.figure(figsize=(8, 8))
plt.title('Proportion of Data by Region', fontsize=16)
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.show()
```

Correlation Heatmap

```python

plt.figure(figsize=(10, 8))
plt.title('Correlation Heatmap', fontsize=16)
corr_matrix = data[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.show()
```

Animated Line Chart for Unemployment Rate

```python

animated_chart = px.line(
    data_frame=data,
    x='Date',
    y='Estimated Unemployment Rate',
    color='Region',
    line_group='States',
    hover_name='States',
    animation_frame='Date',
    title='Animated Unemployment Rate Over Time by Region',
    labels={'Estimated Unemployment Rate': 'Unemployment Rate (%)'},
    template='plotly_dark'
)
animated_chart.update_layout(title_font_size=20, title_x=0.5)
animated_chart.show()
```

Pairplot of Key Features

```python

sns.pairplot(
    data[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'Region']],
    hue='Region',
    palette='coolwarm'
)
plt.show()
```
